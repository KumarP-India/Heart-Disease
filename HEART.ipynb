# Loading Dataset

```python
# Impoerting Lib.(s)
import csv

import pandas as pd


# The Titles of Col.(s)
cols = ['age', 'sex','chest_pain','resting_blood_pressure',  
        'serum_cholestoral', 'fasting_blood_sugar', 'resting_ecg_results',
        'max_heart_rate_achieved', 'exercise_induced_angina', 'oldpeak',"slope of the peak",
        'num_of_major_vessels','thal', 'heart_disease']



data = pd.read_csv('../Datasets/heart.dat', sep = ' ', names=cols) # Loading teh Data to var.

data.sample(frac=1, random_state=42)

data.head() # Showing dataset
```

# Fixing The Dataset

```python
print(data.shape) # The Shape 

print(data.isna().sum()) # Checking for missing Data

print(data.dtypes) # Checking the type

# Replacing 1(False) & 2(True) to 0 & 1(computer binary)
data['heart_disease'] = data['heart_disease'].replace(to_replace = 1, value = 0)
data['heart_disease'] = data['heart_disease'].replace(to_replace = 2, value = 1)

data.head()
```

# X-data

```python
# Importing Lib.(s)
import numpy as np

import matplotlib.pyplot as plt



X = data.drop(columns=['heart_disease']) # Input(s)

data.head()
```

# Y-data

```python
Y = data['heart_disease'].values.reshape(X.shape[0], 1)

shp = X.shape[0]

print(shp)

X = X.to_numpy()

print(X)
```

# Spliting the data

```python
def Split(dt):
    
    trn, tst, val = np.split(dt, [int(.6*len(dt)), int(.8*len(dt))])
    
    return trn, tst, val


X_train, X_Test, X_Val = Split(X)

Y_train, Y_Test, Y_Val = Split(Y)
```

# Defining The structure of Deep Nural Network!

```python
import tensorflow as tf

from tensorflow import keras


DNN = keras.Sequential([
    
    keras.Input(shape=(13,)),
    keras.layers.Dense(200, activation='relu'),
    keras.layers.Dense(168, activation='relu'),
    keras.layers.Dense(333, activation='relu'),
    keras.layers.Dense(84, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
    
])

DNN.summary()
```

# Compiling The Model

```python
DNN.compile(optimizer = 'adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

# Training The Model

```python
ep = [20, 50, 80, 100, 150, 200, 250, 300, 400, 450, 500, 600]

fnl = []

for e in ep:
    Fitting = DNN.fit(X_train, Y_train, epochs=e, validation_data=(X_Val, Y_Val), verbose=1)
    
    print("\n\n")

    result = DNN.evaluate(X_Test, Y_Test)

    print("The accuracy of model is: {}".format(result[1]))
    
    print("\n\n")
    
    fnl.append(result[1])
    
e = ep[fnl.index(max(fnl))]

Fitting = DNN.fit(X_train, Y_train, epochs=e, validation_data=(X_Val, Y_Val), verbose=1)
```

# Evaluating the Final Selected Accuracy

```python
print("\n\n")

result = DNN.evaluate(X_Test, Y_Test)

print("The accuracy of model is: {}",format(result))
```

# Saving the DNN

```python
DNN.save("The-Heart-Disease.h5")
```
